```python
import numpy as np
import pandas as pd
import pylab as pl
import seaborn

%pylab inline
```

```python
secholdings = pd.read_csv('http://www.federalreserve.gov/datadownload/Output.aspx?rel=H41&series=039bdf9cb94977f308f7411d332ceaab&lastobs=&from=01/01/2006&to=10/01/2016&filetype=csv&label=include&layout=seriescolumn')
```

```python
secholdings = secholdings.iloc[5:,:]
secholdings.columns = ['date','total','mbs','tres']
```

```python
pl.plot(pd.to_datetime(secholdings.date), secholdings.total.astype(int)/1000, label = 'Total')
pl.plot(pd.to_datetime(secholdings.date), secholdings.tres.astype(int)/1000, label = 'Treasury')
pl.plot(pd.to_datetime(secholdings.date), secholdings.mbs.astype(int)/1000, label = 'MBS')
pl.legend(loc = 2)
pl.ylabel('Securities Held ($ billions)')
pl.title('Federal Reserve Securities Held Outright')
```

Figure 1: Starting in 2009, the Federal Reserve dramatically increased the size of the System Open Market Account (SOMA) Portfolio in order the support the economy. Now over $4 trillion, the SOMA Portfolio is composed primarily of US Treasury Securities and Morgage Backed Securities (MBS) but also includes smaller amount of Agency Debt and other securities (not pictured individually but included in the "Total" line).
