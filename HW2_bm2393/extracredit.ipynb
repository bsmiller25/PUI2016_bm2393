```python
import os
import pandas as pd
import pylab as pl
import datetime
%pylab inline

assert 'DFDATA' in os.environ
    
dfdata = os.getenv("DFDATA")

waterlevel = pd.read_csv(dfdata + '/zkky-n5j3/1414246141/zkky-n5j3')

waterlevel.head()
```

```python
waterlevel = waterlevel.rename(columns=lambda x: x.strip()) #remove whitespace from column names
waterlevel = waterlevel[['Neversink Date','Neversink Storage']]

waterlevel['Neversink Date'] = pd.to_datetime(waterlevel['Neversink Date'])

waterlevel.head()

```

```python
waterlevel.plot(x = 'Neversink Date', y = 'Neversink Storage')
plt.ylabel("Millions of Gallons")
```
