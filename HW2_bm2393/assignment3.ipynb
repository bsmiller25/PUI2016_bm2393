```python
import os
import pandas as pd
import pylab as pl
%pylab inline

assert 'DFDATA' in os.environ
    
dfdata = os.getenv("DFDATA")

traffic = pd.read_csv(dfdata + '/wng2-85mv/1414246045/wng2-85mv')

traffic.head()


```

```python
traffic = traffic[['2:00-3:00PM','5:00-6:00PM']].dropna()
traffic
```

```python
traffic.plot(x = '2:00-3:00PM', y='5:00-6:00PM',kind = 'scatter')
```
